# **Importing libraries for operations**

```python
import numpy as np #for mathematical operations
import matplotlib.pyplot as plt #for plotting
import pandas as pd #Pandas provides data structures like DataFrames, which are essential for data manipulation, cleaning, and analysis. It enables users to read data from various file formats (e.g., CSV, Excel), process it, and prepare it for machine learning models
import sklearn.datasets # This module offers a collection of sample datasets, both real-world and synthetic, useful for experimenting with machine learning algorithms and practicing data analysis. Examples include the Iris dataset, the Boston housing dataset, and tools for generating artificial datasets.
from sklearn.model_selection import train_test_split # It splits a dataset into two or more subsets for training and testing, allowing users to train a model on one portion of the data and assess its performance on unseen data. This helps prevent overfitting and provides a more realistic estimate of the model's generalization ability.
```

# **Loading the dataset for processing**

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
print(breast_cancer_dataset)
```

# **Loading data to dataframe**

```python
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head() # empty head() shows 5 rows(0 to 4) by default
```

```python
data_frame['target'] = breast_cancer_dataset.target
data_frame.head()
```

# **Splitting dataset for train and test**

```python
X = data_frame.drop(columns=['target'], axis=1)
Y = data_frame['target']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

# **Standerizing the data**

```python
from sklearn.preprocessing import StandardScaler #calculate The mean and standard deviation of each feature.
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train) # fit: Calculates the mean and standard deviation of the training data. transform: Uses those stats to standardize the training data.
X_test_std = scaler.transform(X_test) # Uses the learned parameters (from fit()) to actually change the data. This applies the same transformation (same mean & std) to the test set. You don't fit on the test data, only transform it—so you're not cheating by peeking at test stats.
```

# **Why use fit_transform() on training data and only transform() on test data?**
You want to learn scaling parameters from training data only.

Then apply the same transformation to the test data.

This avoids data leakage (accidentally using test data to influence training).

# **Building Neural Network**

# **Importing libraries for Deep Learning (Neural Network)**

```python
import tensorflow as tf # TensorFlow is an open-source, end-to-end machine learning platform developed by Google. It provides a comprehensive ecosystem of tools, libraries, and resources that allow researchers and developers to build and deploy machine learning models. TensorFlow is particularly well-suited for complex tasks like image recognition, natural language processing, and deep learning, and it can handle large datasets and intricate neural network architectures.
tf.random.set_seed(3) # This is a function in TensorFlow that sets the global random seed to 3. This ensures that the random number generator produces the same sequence of random numbers every time the code is run, as long as the operations depending on it are executed in the same order. This is useful for reproducibility in machine learning experiments, where it is important to be able to get the same results every time the code is run.
from tensorflow import keras # Keras is a high-level API for building and training neural networks. It acts as an interface for other deep learning frameworks, including TensorFlow, and provides a user-friendly way to define and experiment with different neural network layers, models, and optimizers. Keras is designed for ease of use and rapid prototyping, making it suitable for both beginners and experienced practitioners. It simplifies the process of building and training models, and it is often used for tasks such as natural language processing, reinforcement learning, and transfer learning
```

# **Setting up Neural Network Layers**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

# **Description**

**keras.sequential ->** The Keras Sequential model is a way to build neural networks layer by layer. It is suitable for models where the data flows linearly from one layer to the next. This model is created by instantiating the Sequential class and adding layers using the .add() method or by passing a list of layers to the constructor.

**keras.layers.Flatten ->** a high-level deep learning API that reshapes input data into a one-dimensional array, while preserving the batch size.  It’s useful when you need to go from multi-dimensional input (like images or sequences) to a flat vector, typically before passing it to a fully connected (Dense) layer.

**input_shape=(30,) ->**  This specifies the shape of each individual input sample.In this case, it's a 1D vector of 30 elements.

**keras.layers.Dense(20) ->** This means the layer has 20 neurons (or units). Each neuron is connected to all outputs of the previous layer. If the previous layer had, say, 30 values, each of those 30 values will be connected to each of the 20 neurons = 600 weights (plus 20 biases).

**activation='relu' ->** Applies the ReLU (Rectified Linear Unit) activation function to the output of each of the 20 neurons. ReLU just outputs x if x > 0, otherwise 0. It introduces non-linearity, which helps the network learn complex patterns.

**activation='sigmoid' ->** The sigmoid activation is most commonly used in the output layer of binary classification models, because it gives you a probability-like output. The sigmoid function squashes input values into a range between 0 and 1.

# **Compilation of the neural network**

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
    )
```

# **Description**

**model.compile(...) ->** This sets up everything the model needs to train — like choosing a learning strategy (optimizer), the problem type (loss), and how to monitor progress (metrics). Think of it as setting the rules of the game.

**optimizer ->** The optimizer is responsible for updating the weights of the neural network to reduce the error (loss).

**adam ->** (Adaptive Moment Estimation) is a smart, widely used optimization algorithm. It combines ideas from two other optimizers:

*   **Momentum:** remembers the past gradients to smooth out the updates (reduces oscillation).
*   **RMSProp:** adjusts the learning rate for each parameter individually based on how frequently it's updated.

**loss ->** The function tells the model how far off its predictions are from the actual answers. It’s what the optimizer tries to minimize during training.

**Sparse Categorical Crossentropy ->** It’s a loss function used for multi-class classification, where:

*   Your targets (labels) are integers like 0, 1, 2, ... (not one-hot vectors).
*   Your model outputs are a vector of scores (probabilities) for each class.

**metrics ->** This tells Keras what performance metrics to track and show during training.

**Accuracy ->** In multi-class classification, it checks whether the model's highest-probability class matches the true class.
Accuracy is useful only when the classes are balanced. If you have skewed data (e.g., 90% of one class), it can be misleading.
       
       Accuracy = (Number of correct predictions) / (Total predictions)

# **Trainning the Neural Network**

```python
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

# **Description**

**fit() ->** It serves to train a model using the provided training data. During this process, the fit() method analyzes the data and learns the underlying patterns and relationships. For example, in linear regression, it calculates the coefficients for the equation, while in a decision tree, it determines the optimal splits based on features.

**epochs = 100 ->** In the context of training machine learning models, especially neural networks, epochs = 100 means the model will go through the entire training dataset 100 times during training.

# **Visualizing Accuracy & Loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='lower right')
plt.show
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper right')
plt.show
```

# **Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
Y_pred = model.predict(X_test_std)
print(Y_pred)
```

# **Converting the prediction probability to class targets**

```python
Y_pred_labels = [np.argmax(i) for i in Y_pred] #argmax -> gives the index value of the maximum content in a list
print(Y_pred_labels)
```
